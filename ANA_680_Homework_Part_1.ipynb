```python
import pandas as pd
import numpy as np
from sklearn import metrics
import pickle
from sklearn.model_selection import train_test_split
```

```python
white_df = pd.read_csv('winequality-white.csv', sep = ';')
red_df = pd.read_csv('winequality-red.csv', sep = ';')
```

```python
white_df.head()
```

```python
red_df.head()
```

```python
wine_df = pd.concat([white_df, red_df], ignore_index = True)
```

```python
wine_df.head()
```

```python
wine_df.columns = wine_df.columns.str.replace(' ', '_')
```

```python
wine_df.head()
```

```python
wine_df.to_csv('wine_quality.csv', index = False)
```

```python
wine_df.describe()
```

```python
wine_df.corr()
```

```python
X = wine_df.iloc[:, :11]
y = wine_df.iloc[:, 11]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors = 5)
KNN.fit(X_train, y_train)
```

```python
y_pred = KNN.predict(X_test)
```

```python
score = KNN.score(X_test, y_test)
print(score)
```

```python
from sklearn.svm import SVC
```

```python
SVM = SVC(kernel = 'linear', random_state = 0)
SVM.fit(X_train, y_train)
```

```python
y_pred = SVM.predict(X_test)
```

```python
score = SVM.score(X_test, y_test)
print(score)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier(criterion = 'gini', random_state = 1)
DTC.fit(X_train, y_train)
```

```python
y_pred = DTC.predict(X_test)
```

```python
score = DTC.score(X_test, y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier(max_depth = 10, random_state= 0, n_estimators = 10)
RFC.fit(X_train, y_train)
```

```python
score = RFC.score(X_test, y_test)
print(score)
```

```python
with open('RFC_model.pkl', 'wb') as file:
  pickle.dump(RFC, file)
```
